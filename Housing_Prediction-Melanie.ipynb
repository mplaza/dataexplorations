```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

houses = pd.read_csv('../hw/hw2-chicagohouseprices.csv')
```

```python
dummies = pd.get_dummies(houses['HouseType'])
houses = houses.join(dummies)

# dummify house type columns
houses = houses.rename(columns={'Lot/Land': 'Land', 'Multi-Family Home': 'MultiFamily', 'Single-Family Home' : 'SingleFamily'})
```

```python
# extract street name from address and dummify

houses['Address'] = houses['Address'].str.split(r'Ave|St|Pl|Ct|Blvd').str.get(0)
houses['Address'] = houses['Address'].str.split(r'W |N |S |E ').str.get(1).str.strip()
addressdummies = pd.get_dummies(houses['Address'])
houses = houses.join(addressdummies)
houses = houses.rename(columns={'Elk Grove': 'ElkGrove', 'Le Moyne': 'LeMoyne', "Saint Paul": "SaintPaul"})
```

```python
# exclude outliers more than 3 std out of normal price range
pricecutoff = houses['Price'].mean() + 3* houses['Price'].std()
priceoutliers = houses['Price'][houses['Price'] > pricecutoff]
houses = houses.drop(houses.index[priceoutliers.index])
```

```python
# replace na with 0 for bath (gives better fitted results than replacing with mean or 1)
houses['Bath'].fillna(0, inplace=True)
houses.corr()
```

```python
# exclude land that is man std out of range or replace zeroes with mean

landcutoff = houses['LotSizeSqft'].mean() + 3* houses['LotSizeSqft'].std()
landoutliers = houses['LotSizeSqft'][houses['LotSizeSqft'] > landcutoff]
for idx in landoutliers.index:
    houses['LotSizeSqft'][idx] = houses['LotSizeSqft'].mean()

housecutoff = houses['HouseSizeSqft'].mean() + 3* houses['HouseSizeSqft'].std()
houseoutliers = houses['HouseSizeSqft'][houses['HouseSizeSqft'] > housecutoff]
for idx in houseoutliers.index:
    houses['HouseSizeSqft'][idx] = houses['HouseSizeSqft'].mean()
    
landoutliers = houses['LotSizeSqft'][houses['LotSizeSqft'] == 0]
for idx in landoutliers.index:
    houses['LotSizeSqft'][idx] = houses['LotSizeSqft'].mean()
    
houseoutliers = houses['HouseSizeSqft'][houses['HouseSizeSqft'] == 0]
for idx in houseoutliers.index:
    houses['HouseSizeSqft'][idx] = houses['HouseSizeSqft'].mean()
    
```

```python
import statsmodels.formula.api as smf

# see which factors have reasonable p values to max r-squared and adj r squared
lm = smf.ols(formula='Price ~ Age + LotSizeSqft + HouseSizeSqft + Bath + Condo + Land + MultiFamily + Townhome + Augusta + Bell + Damen + Division + ElkGrove + Honore + LeMoyne + Leavitt + Marion + Marshfield + Pierce + Winchester + Wolcott', data=houses).fit()

# print the coefficients
lm.params

lm.summary()
```

```python
# translate the 1342 N Leavitt Street into a dataframe with the app coded vals

X_new = pd.DataFrame({'SingleFamily': 1, 'MultiFamily' : 0, 'Townhome': 0, 'Land': 0, 'Condo': 0, 'Bath': 2, 'Leavitt': 1, 'Augusta' : 0,  'Bell': 0, 'Damen': 0, 'Division': 0, 'ElkGrove' : 0, 'Honore' : 0, 'LeMoyne' : 0, 'Marion': 0, 'Marshfield': 0, 'Pierce' : 0, 'Winchester' : 0, 'Wolcott': 0, 'Age': 116, 'LotSizeSqft': 3049, 'HouseSizeSqft': 3154}, index=[0])
X_new
```

```python
# predict price with lin regression model 

lm.predict(X_new)
```

```python
# compare the linear regression model to the knn based on cross validation and r sqauared scores
```

```python
feature_cols = ['Age', 'LotSizeSqft', 'HouseSizeSqft', 'Bath', 'Condo', 'Land', 'MultiFamily', 'Townhome', 'Augusta', 'Bell', 'Damen', 'Division', 'ElkGrove', 'Honore', 'LeMoyne', 'Leavitt', 'Marion', 'Marshfield', 'Pierce', 'Winchester', 'Wolcott']
X = houses[feature_cols]
y = houses.Price
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.30)
lm = LinearRegression()
lm.fit(X_train, y_train)

lm.score(X_test, y_test)
```

```python
from sklearn.metrics import r2_score
lm = LinearRegression()
lm.fit(X, y)
y_prediction = lm.predict(X)
r2_score(y, y_prediction)
```

```python
from sklearn.cross_validation import cross_val_score, ShuffleSplit

n_samples = len(y)
cv = ShuffleSplit(n_samples, n_iter=10, test_size=0.3)

test_scores = cross_val_score(lm, X, y, cv=cv)
print(test_scores)
test_scores.mean()
```

```python
from sklearn.neighbors import KNeighborsClassifier

myknn = KNeighborsClassifier(1).fit(X_train, y_train)

myknn.score(X_test, y_test)
```

```python
cv = ShuffleSplit(n_samples, n_iter=10, test_size=0.3)

test_scores = cross_val_score(myknn, X, y, cv=cv)
print(test_scores)
test_scores.mean()
```
